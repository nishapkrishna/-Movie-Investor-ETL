```python
# Importing dependencies
import pandas as pd
from sqlalchemy import create_engine
# SQL Alchemy
from sqlalchemy import create_engine
from datetime import datetime
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# Store CSV into dataframe
csv_file = "movie_metadata.csv"
movie_metadata_df = pd.read_csv(csv_file,low_memory=False)
movie_metadata_df.head()
```

```python
# rename "id" column to "movie_id"
movie_metadata_df = movie_metadata_df.rename(columns={'id':'movie_id'})
```

```python
# Copy only the columns we want for the database
bench_movie_metadata_df = movie_metadata_df[['movie_id', 'homepage', 'imdb_id', 'original_title', 'overview','popularity', 'poster_path', 'release_date', 'runtime', 'tagline','title']].copy()

bench_movie_metadata_df.head(3)
```

```python
# loop thru each row to convert entire column of release_date to datetime (strp to breakdown format, then strf to convert
# to string for import to sql database as varchar)
for idx, rows in bench_movie_metadata_df.iterrows():
#     formatted_date = rows['release_date'].stpftime('%Y-%m-%d %H:%M:%S')

    try: 
    
        bench_movie_metadata_df.iloc[idx,7] = (datetime.strptime(rows['release_date'],
                                                                              '%m/%d/%Y' )
                                                                    .strftime('%Y-%m-%d')[:10]
                                                           )
    except:
        
        try:
            bench_movie_metadata_df.iloc[idx,7] = (datetime.strptime(rows['release_date'],
                                                                                  '%Y-%m-%d')
                                                                        .strftime('%Y-%m-%d')[:10]
                                                               )
        except:
            bench_movie_metadata_df = bench_movie_metadata_df.drop(idx, axis=0)
```
